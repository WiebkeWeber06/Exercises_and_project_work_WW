```python
import numpy as np
import cupy as cp

sizes = [128, 256, 512, 1024, 2048]

def test_fft(n, dtype=np.float64):
    print(f"\nSize: {n}x{n}, dtype={dtype}")

    a_np = np.random.random((n, n)).astype(dtype)
    a_cp = cp.asarray(a_np)

    print("NumPy:")
    %timeit np.fft.fft2(a_np)

    print("CuPy:")
    %timeit cp.fft.fft2(a_cp); cp.cuda.Stream.null.synchronize()
```

```python
for n in sizes:
    test_fft(n)
```

```python
for n in sizes:
    test_fft(n, dtype=np.float32)
```

❓ At what sizes does CuPy outperform?

CuPy outperforms NumPy at all tested sizes, starting already from 128×128.


❓ What happens with float32?

When using float32, CuPy becomes even faster and the performance gap to NumPy increases. This is attributed to the fact that GPUs are optimized for single precision (float32).


These results demonstrate that GPU acceleration is highly beneficial for FFT computations, especially for large arrays and lower-precision data types.
